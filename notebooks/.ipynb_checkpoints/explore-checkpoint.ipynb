```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../data/BTX_17-23_Yearly_Totals.xlsx', sheet_name = 0)

df.head(10)
```

```python
#Find out how many NaN or Null are in sheet 0

print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
# Combining all sheets together

all_sheets = pd.read_excel('../data/BTX_17-23_Yearly_Totals.xlsx', sheet_name = None)

print(all_sheets.keys())
```

```python
# Adding the Sheet Name (2017 - 2023) to all the columns

dfs = []

for year, sheet in all_sheets.items():
    sheet['year'] = year
    dfs.append(sheet)

df_all = pd.concat(dfs, ignore_index = True)

print(df_all.shape)
print(df_all.head())
```

```python
type(all_sheets['2017'])
```

```python
# Got an unexpected result from df_all.shape. 
# Found an 's' was dropped from Credit Card BTX, and the WL replaced IP all in 2020

for year, sheet in all_sheets.items():
    print(year, sheet.columns.tolist())
```

```python
# Standardizing Column Names 

standard_columns = {
    'Full Date': 'full_date',
    'Credit Cards BTX': 'cc_btx',
    'Credit Card BTX': 'cc_btx', 
    'Credit Card Supp': 'cc_supp',
    'Credit Card IP': 'cc_wl',
    'Credit Card WL': 'cc_wl',
    'Cash BTX': 'cash_btx',
    'Cash Supp': 'cash_supp',
    'Cash IP': 'cash_wl',
    'Cash WL': 'cash_wl',
    'Check BTX': 'check_btx',
    'Check Supp': 'check_supp',
    'Check IP': 'check_wl',
    'Check WL': 'check_wl',
}

dfs = []

for year, sheet in all_sheets.items():
    sheet = sheet.rename(columns = standard_columns)
    sheet['year'] = year
    dfs.append(sheet)

df_all = pd.concat(dfs, ignore_index = True)

print(df_all.shape)
print(df_all.columns.tolist())
```

The Cells above did the following:  

Identified inconsistent column names across 7 years of data.

Built a mapping to standardize them.

Looped through all sheets and applied it.

Stacked everything into one clean combined dataset.

```python
# Fill NaN values with 0
df_all = df_all.fillna(0)

print(df_all.isnull().sum())
```

```python
print(df_all.dtypes)
print(df_all.head())
print(df_all.tail())
```

```python
# Add useful time columns
df_all['month'] = df_all['full_date'].dt.month
df_all['month_name'] = df_all['full_date'].dt.strftime('%B')
df_all['day_of_week'] = df_all['full_date'].dt.day_name()
df_all['quarter'] = df_all['full_date'].dt.quarter

# Add a total revenue column per day
df_all['total_revenue'] = df_all[['cc_btx', 'cc_supp', 'cc_wl', 
                                    'cash_btx', 'cash_supp', 'cash_wl',
                                    'check_btx', 'check_supp', 'check_wl']].sum(axis=1)

print(df_all[['full_date', 'year', 'month', 'day_of_week', 'total_revenue']].head(10))
```

```python
df_all.to_csv('../data/clinic_revenue_clean.csv', index = False)
```

```python
yearly_revenue = df_all.groupby('year')['total_revenue'].sum()

print(yearly_revenue)
```

```python
df_all.groupby('year')['total_revenue'].describe()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
bars = yearly_revenue.plot(kind='bar', ax=ax, color='steelblue')

# Add value labels on top of each bar
for i, val in enumerate(yearly_revenue):
    ax.text(i, val + 1000, f'${val:,.0f}', ha='center', fontsize=9)

ax.set_title('Annual Revenue 2017-2023')
ax.set_xlabel('Year')
ax.set_ylabel('Total Revenue ($)')
plt.tight_layout()
plt.show()
```

```python
# Find YoY Growth

yoy_growth = yearly_revenue.pct_change() * 100
print(yoy_growth.round(1))
```

## Different Revenue Breakdown over the years ##

```python
# Lets look at BTX Subscription, Supplements, and Weight Loss separately by year

stream_by_year = df_all.groupby('year')[['cc_btx', 'cash_btx', 'check_btx',
                                          'cc_supp', 'cash_supp', 'check_supp',
                                          'cc_wl', 'cash_wl', 'check_wl']].sum()

# Combine payment types into revenue streams
stream_by_year['total_btx'] = stream_by_year['cc_btx'] + stream_by_year['cash_btx'] + stream_by_year['check_btx']
stream_by_year['total_supp'] = stream_by_year['cc_supp'] + stream_by_year['cash_supp'] + stream_by_year['check_supp']
stream_by_year['total_wl'] = stream_by_year['cc_wl'] + stream_by_year['cash_wl'] + stream_by_year['check_wl']

print(stream_by_year[['total_btx', 'total_supp', 'total_wl']])
```

```python
# lets make the above into a pretty pretty picture

stream_by_year[['total_btx', 'total_supp', 'total_wl']].plot(
    kind='bar', 
    stacked=True, 
    figsize=(10,6),
    color=['steelblue', 'coral', 'mediumseagreen']
)

plt.title('Revenue by Stream 2017-2023')
plt.xlabel('Year')
plt.ylabel('Total Revenue ($)')
plt.legend(['BTX', 'Supplements', 'Weight Loss'])
plt.tight_layout()
plt.show()
```

```python
dow_revenue = df_all.groupby('day_of_week')['total_revenue'].mean()
print(dow_revenue)
```

```python
df_all['day_of_month'] = df_all['full_date'].dt.day

dom_revenue = df_all.groupby('day_of_month')['cc_btx'].mean()
print(dom_revenue)
```

```python
# Lets Visualize the numbers above

dom_revenue.plot(kind='bar', figsize=(12, 6), color='steelblue')
plt.title("Avg Daily BTX Revenue by DOM")
plt.xlabel('Day of Month')
plt.ylabel("Average Revenue ($)") 
plt.tight_layout()
plt.show()
```

```python
# How does the payment mix change over time?


payment_by_year = df_all.groupby('year').agg(
    total_cc = ('cc_btx', 'sum'),
    total_cash = ('cash_btx', 'sum'),
    total_check = ('check_btx', 'sum')
).reset_index()

payment_by_year['cc_pct'] = payment_by_year['total_cc'] / (payment_by_year['total_cc'] + payment_by_year['total_cash'] + payment_by_year['total_check']) * 100

print(payment_by_year[['year', 'cc_pct']])
```
